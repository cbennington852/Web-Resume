```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (using pandas for easier handling)
csv = sns.load_dataset("tips")


# Separate features and target variable
X = csv[['total_bill']]  # Feature (independent) - total_bill
y = csv['tip']             # Target (dependent) - tip

# Create a linear regression model
model = sklearn.linear_model.LinearRegression()
model.fit(X , y)
y_preds = model.predict(X)


# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.scatter(X, y, label='Data Points', color='blue')
plt.plot(X, y_preds, color='red', label='Linear Regression')

plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Linear Regression: Total Bill vs. Tip Amount')
plt.legend()
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
csv = pd.read_csv('diamonds.csv')
csv

```

```python
csv['size_class'] = csv['size_class'].round(3).astype(np.float64)
csv['size_class'] = csv['size_class'].replace([np.inf, -np.inf], np.nan)
csv['size_class'].fillna(csv['size_class'].mean(), inplace=True)

csv.to_csv('diamonds.csv')

plt.scatter(csv['price'] , csv['size_class'])

csv
```

```python
X = csv[['price']]  # Feature (independent) - total_bill
y = csv['size_class']     # Target (dependent) - tip

print(max(y))

# Create a linear regression model
model = sklearn.linear_model.LinearRegression()
model.fit(X , y)
y_preds = model.predict(X)
mse = sklearn.metrics.mean_squared_error(y , y_preds)
R2 = sklearn.metrics.r2_score(y, y_preds)


# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.scatter(X, y,)
plt.plot(X, y_preds, color='red', label='Linear Regression')

plt.xlabel('')
plt.ylabel('')
plt.title(f'Linear Regression: MSE:{round(mse , 2)} , R2:{round(R2 , 2)}')
plt.legend()
plt.show()
```

```python
X = csv[['price']]  # Feature (independent) - total_bill
y = csv['size_class']     # Target (dependent) - tip







model = sklearn.tree.DecisionTreeRegressor(
    max_leaf_nodes=3000,
)




model.fit(X , y)
y_preds = model.predict(X)
mse = sklearn.metrics.mean_squared_error(y , y_preds)
R2 = sklearn.metrics.r2_score(y, y_preds)


# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.scatter(X, y,)
plt.scatter(X, y_preds, color='purple', label='DecisionTreeRegressor', s=2)

plt.xlabel('')
plt.ylabel('')
plt.title(f'DecisionTreeRegressor: MSE:{round(mse , 2)} , R2:{round(R2 , 2)}')
plt.legend()
plt.show()
```
