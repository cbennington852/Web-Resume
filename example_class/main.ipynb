```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (using pandas for easier handling)
x_feature = 'total_bill'

csv = sns.load_dataset("tips")
csv = csv.sort_values(x_feature)


# Separate features and target variable
X = csv[[x_feature]]  # Feature (independent) - total_bill
y = csv['tip']             # Target (dependent) - tip

# Create a linear regression model
#model = sklearn.linear_model.LinearRegression()
model = sklearn.tree.DecisionTreeRegressor(
    max_leaf_nodes=12
)
model.fit(X , y)
y_preds = model.predict(X)
mse = round(sklearn.metrics.mean_squared_error(y , y_preds) , 2)
R2 = round(sklearn.metrics.r2_score(y, y_preds) , 2)

# add trans

# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.scatter(X, y, label='Data Points', color='blue')
plt.plot(X, y_preds, color='red', label=f'{model.__class__.__name__}')

plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title(f'{model.__class__.__name__}: Total Bill vs. Tip Amount \nMSE:{mse}\nR2:{R2}')
plt.legend()
plt.grid(True)  # Add a grid for better readability
plt.show()
```

```python
csv = pd.read_csv('diamonds.csv')
csv

```

```python
#csv['shininess'] = csv['size_class']
import random
from sklearn.preprocessing import LabelEncoder
#csv = csv.drop('size_class' , axis=1)
csv['shininess'] = csv['shininess'].round(3).astype(np.float64)
csv['shininess'] = csv['shininess'].replace([np.inf, -np.inf], np.nan)

le = LabelEncoder()
csv['clarity_int'] = le.fit_transform(csv['clarity'])
csv['clarity_int'] = csv['clarity_int'].apply(lambda x : x + (random.randrange(0,1000)/1000))
csv['clarity'] = csv['clarity_int']
csv['shininess'].fillna(csv['shininess'].mean(), inplace=True)
csv = csv.drop('clarity_int' , axis=1)

csv.to_csv('diamonds.csv')

plt.scatter(csv['clarity'] , csv['price'])
csv
```

```python
X = csv[['price']]  # Feature (independent) - total_bill
y = csv['size_class']     # Target (dependent) - tip

print(max(y))

# Create a linear regression model
model = sklearn.linear_model.LinearRegression()
model.fit(X , y)
y_preds = model.predict(X)
mse = sklearn.metrics.mean_squared_error(y , y_preds)
R2 = sklearn.metrics.r2_score(y, y_preds)


# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))  # Adjust figure size
plt.scatter(X, y,)
plt.plot(X, y_preds, color='red', label='Linear Regression')

plt.xlabel('')
plt.ylabel('')
plt.title(f'Linear Regression: MSE:{round(mse , 2)} , R2:{round(R2 , 2)}')
plt.legend()
plt.show()
```
